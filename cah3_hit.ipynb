```python
from Bio.SeqUtils.ProtParam import ProteinAnalysis
# https://biopython.org/docs/1.76/api/Bio.SeqUtils.ProtParam.html

# df = pd.read_csv('CAH3_Interactor_Seq_2.csv')
# display(df)
import striprtf.striprtf
from striprtf.striprtf import rtf_to_text

with open("cah3_interactor_seq.rtf") as infile:
    content = infile.read()
    text = rtf_to_text(content)

seqs = text.split('>')
# print(seqs)

# seqs_mini = seqs[1]
# seqs_mini_list = seqs_mini.split('\n')
# print(seqs_mini_list)

id_list = []
aa_seq_list = []
seqs_mini_list = []

for y in range(0, len(seqs)):
    if seqs[y] != '':
        seqs_mini_list = seqs[y].split('\n')
        # print(seqs_mini_list)
        id_list.append(seqs_mini_list[0].split('|')[1])
        if '*' in seqs_mini_list[1]:
            aa_seq_list.append(seqs_mini_list[1].split('*')[0])
        else:
            aa_seq_list.append(seqs_mini_list[1].split('|')[0])

'''print(len(id_list))
print(len(aa_seq_list))
print(id_list)
print(aa_seq_list)'''
```

```python
def molecular_weight(seq_list):
    mw_list = []
    for y in range(0, len(seq_list)):
        X = ProteinAnalysis(seq_list[y])
        mw_list.append(float(("%0.2f" % X.molecular_weight())))
    return mw_list
mws = molecular_weight(aa_seq_list)

data = {'Cre ID': id_list, 'Molecular Weight': mws, 'Sequence': aa_seq_list}
df = pd.DataFrame(data=data)
# display(df)

sorted_df_asc = df.sort_values(by='Molecular Weight')
display(sorted_df_asc)
```

```python
def make_tiles(string_name, j_val):
    test_list = list(string_name)
    
    #5 character string, shifted by 1 every time 
    substring = ''
    substring_list = []
    
    i = 0
    j = j_val
    while j <= len(test_list):
        # print(i)
        # print(j)
        for x in range(i, j):
            substring += test_list[x]
        substring_list.append(substring)
        i += 1
        j += 1
        substring = ''
    
    return substring_list

        
# test = 'MFTAVNSNPNASESISGNSAFNFPSAPVSSLDTNNYGQRRPSLLSGTSPTSSFFNSSMISSNYTFPHGSNKQASLESPVSYSNPIPSLTWLSLDGDSPDSLVSTPTAPSA'
# make_tiles(test, 5)

five_til = []
ten_til = []

for x in range(0, len(aa_seq_list)):
    three_til = make_tiles(aa_seq_list[x], 3)
    four_til = make_tiles(aa_seq_list[x], 4)
    five_til = make_tiles(aa_seq_list[x], 5)
    six_til = make_tiles(aa_seq_list[x], 6)
    # ten_til = make_tiles(aa_seq_list[x], 10)
    current_seq = aa_seq_list[x]
    
    three_list_of_lists = []
    four_list_of_lists = []
    five_list_of_lists = []
    six_list_of_lists = []
    ten_list_of_lists = []
    
    for a in range(0, len(three_til)):
        new_list = []
        for b in range(0, len(aa_seq_list)):
            if three_til[a] in aa_seq_list[b]:
                if current_seq != aa_seq_list[b]:
                    new_list.append(three_til[a] + '-' + aa_seq_list[b])
        if len(new_list) != 0:
            three_list_of_lists.append(new_list)
    
    for c in range(0, len(four_til)):
        new_list = []
        for d in range(0, len(aa_seq_list)):
            if four_til[c] in aa_seq_list[d]:
                if current_seq != aa_seq_list[d]:
                    new_list.append(four_til[c] + '-' + aa_seq_list[d])
        if len(new_list) != 0:
            four_list_of_lists.append(new_list)
    
    
    for y in range(0, len(five_til)):
        new_list = []
        for z in range(0, len(aa_seq_list)):
            if five_til[y] in aa_seq_list[z]:
                if current_seq != aa_seq_list[z]:
                    new_list.append(five_til[y] + '-' + aa_seq_list[z])
        if len(new_list) != 0:
            five_list_of_lists.append(new_list)
                                              
    for e in range(0, len(six_til)):
        new_list = []
        for f in range(0, len(aa_seq_list)):
            if six_til[e] in aa_seq_list[f]:
                if current_seq != aa_seq_list[f]:
                    new_list.append(six_til[e] + '-' + aa_seq_list[f])
        if len(new_list) != 0:
            six_list_of_lists.append(new_list)
    
'''for p in range(0, len(ten_til)):
new_list_1 = []
for q in range(0, len(aa_seq_list)):
    if ten_til[p] in aa_seq_list[q]:
        if current_seq != aa_seq_list[q]:
            new_list_1.append(ten_til[p] + '-' + aa_seq_list[q])
if len(new_list) != 0:
    ten_list_of_lists.append(new_list_1)'''

# print(len(five_list_of_lists)) 
# print(len(ten_list_of_lists)) 
# print(five_list_of_lists)

new_list = []
all_seqs = []
seq_count = []
motif_list = []

def make_df(list_name):
    new_list = []
    all_seqs = []
    seq_count = []
    motif_list = []
    for x in range(0, len(list_name)):
        seq_list = []
        for y in range(0, len(list_name[x])):
            new_list = list_name[x][y].split('-')
            motif = new_list[0]
            seq_list.append(new_list[1])
        if len(seq_list) > 1:
            motif_list.append(motif)
            all_seqs.append(seq_list)
            seq_count.append(len(seq_list))

    data = {'Motif': motif_list, 'Instance Count': seq_count}
    motif_instance = pd.DataFrame(data=data)
    sorted_df_asc = motif_instance.sort_values(by='Instance Count')
    # display(sorted_df_asc)
    return sorted_df_asc
    

print('3 AA Motif')
new_df = make_df(three_list_of_lists)
display(new_df.loc[[17, 85, 21, 18, 51, 19, 20]])
print('4 AA Motif')
new_df_1 = make_df(four_list_of_lists)
display(new_df_1.loc[[21, 8, 5, 7, 6]])
print('5 AA Motif')
new_df_2 = make_df(five_list_of_lists)
display(new_df_2)
print('6 AA Motif')
new_df_3 = make_df(six_list_of_lists)
display(new_df_3)
```
